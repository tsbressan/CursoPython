```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
import numpy as np
```

```python
#nome do arquivo
file = "gra_10_9_2019.csv"

#ler arquivos csv com delimitador informado
dataset = pd.read_csv(file,sep=",")
```

Separa a quantidade de registros por Site - cria uma s√©rie(pandas.core.series.Series)
Matriz unidirecional com label por coluna


```python
quantidade = dataset['Site'].value_counts()
```

```python
type(quantidade)
```

Exemplo de loop mostrando o label e o valor

```python
labels = []
valor = []
```

```python
for label, val in quantidade.iteritems():
    labels.append (label)
    valor.append (val)
    print (label, val)
```

```python


x = np.arange(len(labels))  # the label locations
width = 0.50  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, valor, width, label='Site')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Total number of records by site')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
#autolabel(rects2)

fig.tight_layout()

fig.savefig("teste.png")
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:

explode = (0.1, 0.1, 0, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(valor, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax1.legend(labels=labels,title="Legend:",
          loc="center left",
          bbox_to_anchor=(0.95, 0, 1, 1),prop={'size': 14})

ax1.set_title("Total number of records by site", fontsize=16)

plt.show()
```
